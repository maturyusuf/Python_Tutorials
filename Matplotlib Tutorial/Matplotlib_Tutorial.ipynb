```python
import random
from itertools import count
import pandas as plt
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use("fivethirtyeight")

x_vals = []
y_vals = []

index = count() # This function is used for counting values.It adds up to its value everytime it's triggered .

def animate(i):
  x_vals.append(next(count)) # "next(count)" function will add upp to the index value
  y_vals.append(random.randint(0,5))

  plt.cla()  # "plt.cla()" Clears axis and prevent figures to overlap by of FuncAnimation method 
  plt.plot(x_vals,y_vals)


ani=FuncAnimation(plt.gcf(),animate,interval=1000)    #FuncAnimation(["current figure"],["function to animate"],["animation loop time (ms)"])


plt.tight_layout()
plt.show()
```
